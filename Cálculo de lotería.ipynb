```python
import numpy as np
import pandas as pd
from pandas import DataFrame
from itertools import product

#recoge los elementos de la página web, lo vuelve texto y extrae los números
import urllib.request
pagina = urllib.request.urlopen("https://www.astroluna.co/bogota").read().decode() #para cualquier lotería solo es cambiar la url en esta misma página
paginaweb=pagina[pagina.find("4a</")+170:pagina.find("Sorteos sin salir de cada")-385]
paginaweb

import re

lista_numeros=list(([int(s) for s in re.findall(r'-?\d+\.?\d*', paginaweb)]))

#parte la lista en grupos de 10 y organiza para visualizar entradas de procedimientos de análisis

x = 8

final_list= lambda lista_numeros, x: [lista_numeros[i:i+x] for i in range(0, len(lista_numeros), x)]

output=final_list(lista_numeros, x)

output=DataFrame(output)
output.iloc[:,0]
primera=list(output.iloc[:,0])
c_primera=list(output.iloc[:,1])
segunda=list(output.iloc[:,2])
c_segunda=list(output.iloc[:,3])
tercera=list(output.iloc[:,4])
c_tercera=list(output.iloc[:,5])
cuarta=list(output.iloc[:,6])
c_cuarta=list(output.iloc[:,7])

#arma el dataframe y cambia de nombres a las columnas

datos=[primera,c_primera,segunda,c_segunda,tercera,c_tercera,cuarta,c_cuarta]

df_datos=DataFrame(datos).transpose()
nombres=["primera","c_primera","segunda","c_segunda","tercera","c_tercera","cuarta","c_cuarta"]
df_datos.columns=nombres

#extrae cada cifra y calcula su ratio de salida
total=df_datos["c_primera"].sum()
p_primera=[]
for i in df_datos["c_primera"]:
    p=i/total
    p_primera.append(p)
df_datos["p_primera"]=p_primera

total=df_datos["c_segunda"].sum()
p_segunda=[]
for i in df_datos["c_segunda"]:
    p=i/total
    p_segunda.append(p)
df_datos["p_segunda"]=p_segunda


total=df_datos["c_tercera"].sum()
p_tercera=[]
for i in df_datos["c_tercera"]:
    p=i/total
    p_tercera.append(p)
df_datos["p_tercera"]=p_tercera


total=df_datos["c_cuarta"].sum()
p_cuarta=[]
for i in df_datos["c_cuarta"]:
    p=i/total
    p_cuarta.append(p)
df_datos["p_cuarta"]=p_cuarta

#crea df para cada cifra: número con ratio
df_datos=df_datos.drop(["c_primera","c_segunda","c_tercera","c_cuarta"],axis=1)
d1=df_datos[["primera","p_primera"]].sort_values("primera")
d1=d1["p_primera"]
d2=df_datos[["segunda","p_segunda"]].sort_values("segunda")
d2=d2["p_segunda"]
d3=df_datos[["tercera","p_tercera"]].sort_values("tercera")
d3=d3["p_tercera"]
d4=df_datos[["cuarta","p_cuarta"]].sort_values("cuarta")
d4=d4["p_cuarta"]


#multiplica cada p_primera por cada p_segunda
d1_2=[]
for i in d1:
    for j in d2:
        k=float(i*j)
        d1_2.append(k)

#multiplica cada p_segunda por cada p_tercera
d2_3=[]
for m in d1_2:
    for l in d3:
        k=float(m*l)
        d2_3.append(k)
        
#multiplica cada p_tercera por cada p_cuarta
d3_4=[]
for n in d2_3:
    for d in d4:
        k=float(n*d)
        d3_4.append(k)


#ya tenemos los ratios de todos los números de 0000 a 9999

#crea todos los números posibles y junta con las ratios
x = "0123456789"
todos=list(product(x, repeat=4))
final=DataFrame(d3_4) #no puede meter la otra porque cree que está haciendo index con "todos"
final["número"]=todos

#termina marcando cuáles son los números que menos han salido y deberían subir
nombres=["ratio","número"]
final.columns=nombres
final.sort_values(["ratio"],ascending=True,axis=0,inplace=True) #si usted quiere saber cuáles son los que más ratio tiene ascendgin=False
final=final.head(20)
print(final)
```
